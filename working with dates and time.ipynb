#  Working with Dates

```python
# # Converting to datetime datatype
date['date']= pd.to_datetime(date['date'])
```

```python
date.info
```

```python
## extract yeat
```

```python
date['date_year'] = date['date'].dt.year
```

```python
date.sample(5)
```

```python
# extract month
```

```python
date['date_month_no'] = date['date'].dt.month
date.head()
```

```python
date['date_month_name']= date['date'].dt.month_name()

date.head()
```

```python
# extract days

date['date_day']= date['date'].dt.day

date.head()
```

```python
##extract day

date['date_day'] = date['date'].dt.dayofweek

date.head()
```

```python
# day of week name

date['date_day_name']= date['date'].dt.day_name()

date.head()
```

```python
## is it a weekand

date['date_is_weekand']= np.where(date['date_day_name'].isin(['Sunday','Saturday']),1,0)
date.head()
```

```python
date.drop(columns=['product_id','city_id','orders'])
```

```python
## Extract week of year

date['date_week'] = date['date'].dt.week

date.head()
```

```python
date['date_week'] = date['date'].dt.week

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
## Extract quater

date['quarter'] = date['date'].dt.quarter

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# Extract Semester
date['sem']= np.where(date['quarter'].isin([1,2]),1,2)
```

```python
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# in hours

(today - time['date'])/np.timedelta64(1,'h')
```
