# Box Plot(Numerical-categorical)

```python
#xaxis is for categorical data and y is for numerical data
```

```python
sns.boxplot(data=df,x="Sex",y="Age")
```

```python
#outlier is present in male (the black dot)
```

```python
sns.boxplot(data=df,x="Sex",y='Age',hue="Survived")
```

# Distplot(Numerical-Categorical)

```python
sns.distplot(df['Age'])
```

```python
sns.distplot(df[df["Survived"]==0]["Age"])
```

```python
#this is pdf of people who dont survived
```

```python
sns.distplot(df[df["Survived"]==0]["Age"],hist=False)
sns.distplot(df[df["Survived"]==1]["Age"],hist=False)
```

```python
## orange=p(survived) and P(not survived)= Blue
##for the age below 0 probobility of surviving is more than the probability of 
#dying
```

```python
## from the age 15-30 P(dying) is more
```
